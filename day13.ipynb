```python
# 58. Reverse a list using slicing and using .reverse() — compare the two approaches.
# 1. Using .reverse() (Changes the Original) 🔄
num =[1,2,3,4,5] 
num.reverse()
print(num)

# using slicing method
# 2. Using [::-1] (Keeps Original Safe) 🛡️
num =[1,2,3,4,5]
reverse =num[::-1]
print(reverse)
```

```python
# 59. Sort a list in ascending and descending order using .sort() and sorted() .
name =['virat','rohith','abhishek','rahul','dinesh']
name.sort() 
# by default ascending order
# changes the orignal copy 
print(name)
```

```python
num =[1,254,7,8,9,6]
print(name2:=sorted(num))
# remains the original copy
# descnding order

num =[1,254,7,8,9,6]
# desending order
num.sort(reverse=True)
print(num)

# can also do kind
num =[1,254,7,8,9,6]
num.sort()
print(num2:=num[::-1])

```
